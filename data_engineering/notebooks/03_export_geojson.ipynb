```python
import pandas as pd
import geopandas as gpd 
from shapely.geometry import Point
```

```python
df_clean = pd.read_csv("../data/clean/riyadh_restaurants_clean.csv")
```

```python
geometry = [Point(xy) for xy in zip(df_clean["lng"] , df_clean["lat"])]
```

```python
gdf_rest = gpd.GeoDataFrame(
    df_clean,
    geometry=geometry,
    crs = "EPSG:4326"
)
```

```python
gdf_rest.to_file("../data/clean/riyadh_restaurants_clean.geojson", driver="GeoJSON")
```
